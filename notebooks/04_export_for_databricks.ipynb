# 04 — Export Data for Databricks

Exports processed parquet files to CSV format for upload to Azure Databricks Unity Catalog.

```python
import pandas as pd
from pathlib import Path

processed_dir = Path('../data/processed')
exports_dir = Path('../data/exports')
exports_dir.mkdir(parents=True, exist_ok=True)
```

## 1. Export Medical Chunks

```python
chunks_df = pd.read_parquet(processed_dir / 'medical_chunks.parquet')
chunks_df.to_csv(exports_dir / 'medical_chunks.csv', index=False)
print(f'Exported {len(chunks_df)} chunks to medical_chunks.csv')
print(f'Columns: {list(chunks_df.columns)}')
```

## 2. Export Eval Queries

```python
eval_df = pd.read_parquet(processed_dir / 'eval_queries.parquet')
eval_df.to_csv(exports_dir / 'eval_queries.csv', index=False)
print(f'Exported {len(eval_df)} eval queries to eval_queries.csv')
```

## 3. Export Test Queries

```python
test_df = pd.read_parquet(processed_dir / 'test_queries.parquet')
test_df.to_csv(exports_dir / 'test_queries.csv', index=False)
print(f'Exported {len(test_df)} test queries to test_queries.csv')
```

## 4. Summary

```python
csv_files = list(exports_dir.glob('*.csv'))
total_size = sum(f.stat().st_size for f in csv_files)
print(f'Total CSV files: {len(csv_files)}')
print(f'Total size: {total_size / 1024:.0f} KB')
for f in sorted(csv_files):
    print(f'  {f.name}: {f.stat().st_size / 1024:.0f} KB')
```
